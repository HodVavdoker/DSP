```python

csv_file = 'DSPHT'
target_words = ['סיני','חלב','בשר','הזמנה','סושי','בורגר','משלוח','קפה','דג' ,'גריל' ,'תאילנדי' ,'איטלקי','אסייתית' ]
final=''
for index, target_word in enumerate(target_words):
    print(target_word+' is added')
    # Read the CSV file into a pandas DataFrame
    if index == 0:
        df = pd.read_csv(csv_file+'.csv')
    else:
        df = pd.read_csv(csv_file+str(index-1)+'.csv')

    # Merge the columns based on the target word
    merged_columns = df.filter(like=target_word).max(axis=1)

    # Keep the non-merged columns
    non_merged_columns = df.drop(columns=df.filter(like=target_word).columns)

    # Concatenate the merged and non-merged columns
    merged_df = pd.concat([merged_columns, non_merged_columns], axis=1)

    # Assign the target word as the column header
    merged_df.columns = [target_word] + non_merged_columns.columns.tolist()

    # Write the merged DataFrame to a new CSV file
    if index==len(target_words)-1:
        final='_FINAL'
    merged_filename = csv_file+str(index)+final+'.csv'
    merged_df.to_csv(merged_filename, encoding='utf-8-sig', index=False, header=target_word)

print(f"the last Merged columns doc have been written to {merged_filename}")

```
